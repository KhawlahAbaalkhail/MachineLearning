```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("creditcard.csv")
df.head()
```

```python
df.shape
df.info()
df.describe()
df.isnull().sum()
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
plt.figure()
plt.hist(df["Amount"], bins=30)
plt.title("Distribution of Transaction Amount")
plt.xlabel("Amount")
plt.ylabel("Frequency")
plt.show()
```

The distribution of transaction amounts is highly right-skewed, indicating most transactions involve small amounts.


```python
plt.figure()
plt.hist(df["Time"], bins=30)
plt.title("Distribution of Transaction Time")
plt.xlabel("Time")
plt.ylabel("Frequency")
plt.show()
```

Transactions are spread across time, showing continuous activity throughout the dataset.


```python
plt.figure()
df["Class"].value_counts().plot(kind="bar")
plt.title("Fraud vs Non-Fraud Transactions")
plt.xlabel("Class (0 = Non-Fraud, 1 = Fraud)")
plt.ylabel("Count")
plt.show()
```

The dataset is highly imbalanced, with fraud cases representing a very small percentage of total transactions.


```python
plt.figure()
sns.boxplot(x="Class", y="Amount", data=df)
plt.title("Transaction Amount by Class")
plt.show()
```

Fraud transactions tend to have different amount distributions compared to non-fraud transactions.


```python
plt.figure()
sns.boxplot(x="Class", y="Time", data=df)
plt.title("Transaction Time by Class")
plt.show()
```

Fraudulent and non-fraudulent transactions show variation across time.


```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

Most features show low correlation, which is expected due to anonymized PCA components.
